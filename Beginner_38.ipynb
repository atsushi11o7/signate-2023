```python
import xgboost as xgb
from xgboost import XGBClassifier
```

```python
model_XGB = XGBClassifier(booster="gbtree",             # gbtree or dart, gblinear)
                          learning_rate=0.0001,              # 
                          min_split_loss=1,             # 
                          max_depth=1,                  # 
                          min_child_weight=0.1,           # 
                          subsample=0.5,                  # 
                          sampling_method="uniform",    # 
                          colsample_bytree=0.5,           # 
                          colsample_bylevel=1,          # 
                          reg_lambda=0,                 # L2
                          reg_alpha=0,                  # L1
                          tree_method="auto",           # 
                          process_type="default",       # 
                          grow_policy="depthwise",      # 
                          max_leaves=2,                 # 
                          objective="reg:squarederror", # 
                          #early_stopping_rounds = 10,                  # (=)
                          n_estimators=100000,
                         )
```

```python
# 
model_XGB.fit(X_train, Y_train)
```

```python
Y_pred_XGB = model_XGB.predict(X_test)

# 
print(Y_pred_XGB)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score

# 
Y_pred_XGB = model_XGB.predict(X_test)

# 
print(": " + str(round(accuracy_score(Y_test,Y_pred_XGB),3)))
print(": " + str(round(precision_score(Y_test,Y_pred_XGB, average="macro"),3)))
print(": " + str(round(recall_score(Y_test,Y_pred_XGB, average="macro"),3)))

# 
# : 0.978
# : 0.978
# : 0.976
```
