```python
import numpy as np
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split
```

```python
train_data = pd.read_csv('/content/drive/My Drive/Colab Notebooks/data/train.csv')
test_data = pd.read_csv('/content/drive/My Drive/Colab Notebooks/data/test.csv')
train_data.head()
```

```python
test_data.head()
```

```python
unique_values_grade = train_data['grade'].unique()
unique_values_purpose = train_data['purpose'].unique()
unique_values_application_type = train_data['application_type'].unique()

print(unique_values_grade)
print(unique_values_purpose)
print(unique_values_application_type)
```

```python
unique_values_grade = test_data['grade'].unique()
unique_values_purpose = test_data['purpose'].unique()
unique_values_application_type = test_data['application_type'].unique()

print(unique_values_grade)
print(unique_values_purpose)
print(unique_values_application_type)
```

```python
# apply() メソッドを使って置き換える
train_data['term'] = train_data['term'].apply(lambda x: x.split()[0] if 'years' or 'year' in x else x)
test_data['term'] = test_data['term'].apply(lambda x: x.split()[0] if 'years' or 'year' in x else x)

train_data['employment_length'] = train_data['employment_length'].apply(lambda x: x.split()[0] if 'years' or 'year' in x else x)
test_data['employment_length'] = test_data['employment_length'].apply(lambda x: x.split()[0] if 'years' or 'year' in x else x)

# 数値型に変換
train_data['term'] = train_data['term'].astype(int)
test_data['term'] = test_data['term'].astype(int)

train_data['employment_length'] = train_data['employment_length'].astype(int)
test_data['employment_length'] = test_data['employment_length'].astype(int)


train_data = train_data.replace({'A1': 1, 'A2': 2, 'A3': 3, 'A4': 4, 'A5': 5,
                                 'B1': 6, 'B2': 7, 'B3': 8, 'B4': 9, 'B5': 10,
                                 'C1': 11, 'C2': 12, 'C3': 13, 'C4': 14, 'C5': 15,
                                 'D1': 16, 'D2': 17, 'D3': 18, 'D4': 19, 'D5': 20,
                                 'E1': 21, 'E2': 22, 'E3': 23, 'E4': 24, 'E5': 25,
                                 'F3': 26, 'F5': 27})
test_data = test_data.replace({'A1': 1, 'A2': 2, 'A3': 3, 'A4': 4, 'A5': 5,
                               'B1': 6, 'B2': 7, 'B3': 8, 'B4': 9, 'B5': 10,
                               'C1': 11, 'C2': 12, 'C3': 13, 'C4': 14, 'C5': 15,
                               'D1': 16, 'D2': 17, 'D3': 18, 'D4': 19, 'D5': 20,
                               'E1': 21, 'E2': 22, 'E3': 23, 'E4': 24, 'E5': 25,
                               'F3': 26, 'F5': 27})

train_data = train_data.replace({'debt_consolidation': 0, 'credit_card': 1, 'home_improvement': 2,
                                 'medical': 3, 'car': 4, 'major_purchase': 5,
                                 'small_business': 6, 'house': 7, 'other': 8})
test_data = test_data.replace({'debt_consolidation': 0, 'credit_card': 1, 'home_improvement': 2,
                               'medical': 3, 'car': 4, 'major_purchase': 5,
                               'small_business': 6, 'house': 7, 'other': 8, 'moving': 9})

train_data = train_data.replace({'Individual': 0, 'Joint App': 1})
test_data = test_data.replace({'Individual': 0, 'Joint App': 1})

train_data = train_data.replace({'FullyPaid': 0, 'ChargedOff': 1})


train_data.head()
```

```python
# 学習データとテストデータ分割
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size=0.2, shuffle=True, random_state=3)
```

```python
X_train = X
Y_train = Y
X_test = test_data
```

```python
import xgboost as xgb
from xgboost import XGBClassifier
```

```python
model_XGB = XGBClassifier(booster="gbtree",             # ブースター種類（ツリーモデル：gbtree or dart, 線形モデル：gblinear)
                          learning_rate=0.01,              # 過学習防止を目的とした学習率
                          min_split_loss=1,             # 決定木の葉ノード追加に伴う損失減少の下限値
                          max_depth=1,                  # 決定木の深さの最大値
                          min_child_weight=0.1,           # 決定木の葉に必要な重みの下限
                          subsample=0.5,                  # 各決定木においてランダム抽出されるサンプル割合
                          sampling_method="uniform",    # サンプリング方法
                          colsample_bytree=0.5,           # 各決定木でランダムに設定される説明変数の割合
                          colsample_bylevel=1,          # 決定木が深さ単位で分割される際に利用する説明変数の割合
                          reg_lambda=0,                 # L2正則化のペナルティ項
                          reg_alpha=0,                  # L1正則化のペナルティ項
                          tree_method="auto",           # ツリー構造アルゴリズム
                          process_type="default",       # 実行するブースティングプロセス
                          grow_policy="depthwise",      # 新しい葉ノードを木に追加する際の制御ポリシー
                          max_leaves=2,                 # 追加する葉ノードの最大数
                          objective="reg:squarederror", # 学習プロセスで最小化を目指す損失関数
                          #early_stopping_rounds = 10,                  # ブースティング回数(=作成する決定木の本数)
                          n_estimators=1000,
                         )
```

```python
# モデル学習
model_XGB.fit(X_train, Y_train)
```

```python
Y_pred_XGB = model_XGB.predict(X_test)

# 出力
print(Y_pred_XGB)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score

# 推論
Y_pred_XGB = model_XGB.predict(X_test)

# 性能評価
print("正解率: " + str(round(accuracy_score(Y_test,Y_pred_XGB),3)))
print("適合率: " + str(round(precision_score(Y_test,Y_pred_XGB, average="macro"),3)))
print("再現率: " + str(round(recall_score(Y_test,Y_pred_XGB, average="macro"),3)))
```

```python
pred_df = pd.DataFrame(Y_pred_XGB)
output = pd.concat([id, pred_df], axis=1)

output.head(20)
```
